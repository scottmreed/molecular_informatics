Overview
You will use MolModa to perform molecular docking experiments on muscle glycogen phosphorylase (PDB ID: 1c8k) and ligand molecules from Assignment 8. Ensure you are using Chrome, as this method does not work on Safari.

Steps to Complete the Assignment:
1	Retrieve Ligand List
◦	Re-run 8_virtual_screening.ipynb to retrieve the list of 5 values in my_cids.

2	Open MolModa
◦	Access MolModa at https://durrantlab.pitt.edu/molmoda/.

3	Load the Ligand Structure
◦	Use File > Import Compound > PubChem to import from the CID number. Select "desalt".
◦	In the Navigator, select the molecule, then protonate it at pH 7.4 using the Protonate tool.

◦	Delete the original molecule after protonation.
4	Load the Protein Structure
◦	Go to File > PDB ID and enter 1c8k. The structure, including all chains, will load.

5	Prepare the Target Protein
◦	In the Navigator pane:
▪	Find, select, and delete solvent molecules.
▪	Find, select, and delete the two co-crystallized molecules.

6	Protonate the Protein
◦	Select Proteins > Protonation to add hydrogen atoms.
◦	Delete the original unprotonated protein.

7	Identify Binding Pockets
◦	Use Detect Pockets to locate binding sites.
◦	In the Navigator pane, focus on the inhibitor CPB.
◦	Click the eyeball icon on each pocket until you find one that overlaps with CPB.

8	Delete the Ligand
◦	In the Navigator pane, delete the CPB ligand.
9	Set Up the Docking Experiment
◦	Go to Docking > Compound.
◦	Select the target pocket and the prepared ligand.
◦	Adjust the docking parameters (e.g., exhaustiveness) as needed.

10	Run and Review Docking Results
◦	Execute the docking run.
◦	Review the poses and scores.
◦	Visualize top poses and analyze interactions.

11	Repeat for Additional Ligands
◦	Load and dock other ligand molecules using the same procedure.
◦	Save and analyze the docking results.
